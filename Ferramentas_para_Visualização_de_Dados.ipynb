####Estatistica Descritiva - FEMININO

#####Pontos Médios

-> Para simplificar o cálculo de estatísticas descritivas, as idades foram reprsentadas pelos pontos médios de cada faixa etária. Essa escolha facilta os calculos subsequentes.

```python
pontos_medios_fem = pd.DataFrame(grouped_df_fem2)
pontos_medios_fem
```

```python
#dicionário para mapear as faixas etárias para valores numéricos
faixa_etaria_map = {
    '16 anos': 16,
    '17 anos': 17,
    '18 a 20 anos': 19,  # Valor médio da faixa
    '21 a 24 anos': 22.5,  # Valor médio da faixa
    '25 a 34 anos': 29.5,  # Valor médio da faixa
    '35 a 44 anos': 39.5,  # Valor médio da faixa
    '45 a 59 anos': 52,  # Valor médio da faixa
    '60 a 69 anos': 64.5,  # Valor médio da faixa
    '70 a 79 anos': 74.5,  # Valor médio da faixa
    'Superior a 79 anos': 80  # Valor representativo
}
```

```python
# Substituir as faixas etárias pelos valores numéricos
pontos_medios_fem['faixa_etaria'] = pontos_medios_fem['DS_FAIXA_ETARIA'].map(faixa_etaria_map)
```

```python
pontos_medios_fem
```

#####Média

```python
# Acrescentando a coluna 'multp', que é a multiplicação da coluna 'faixa_etaria' com 'counts'
pontos_medios_fem['multp'] = pontos_medios_fem['faixa_etaria']*pontos_medios_fem['counts']
pontos_medios_fem
```

```python
# Calculando a divisão da soma total de 'multp' pela soma total de 'counts'
media_fem = pontos_medios_fem['multp'].sum() / pontos_medios_fem['counts'].sum()

print(media_fem)
```

#####Mediana

```python
# Ordenar o dataframe pela coluna 'counts'
pontos_medios_fem_sorted = pontos_medios_fem.sort_values(by='counts').reset_index(drop=True)
pontos_medios_fem_sorted
```

```python
# Como são 10 faixas (par)  o valor da mediana é baseado nas faixas etárias que estão nas posições 4 e 5
# Fazer calculo dos dados dos index 4 e 5
soma_multp1 = pontos_medios_fem_sorted.loc[4, 'multp'] + pontos_medios_fem_sorted.loc[5, 'multp']

# Fazer a soma dos dados dos index 4 e 5 na coluna 'counts'
soma_counts1 = pontos_medios_fem_sorted.loc[4, 'counts'] + pontos_medios_fem_sorted.loc[5, 'counts']

# Dividir a soma da coluna 'multp' pela soma da coluna 'counts'
mediana = soma_multp1 / soma_counts1

mediana
```

#####Desvio Padrão

Indica o quão dispersas as idades estão em relação a média. Quanto maior o valor maior a variabilidade nas idades.

```python
# Calcular o quadrado da diferença de cada valor em relação à média
pontos_medios_fem_sorted['desvio_p'] = (pontos_medios_fem_sorted['faixa_etaria'] - media_fem) ** 2
# Transformar a coluna 'desvio_p' em int
pontos_medios_fem_sorted['desvio_p'] = pontos_medios_fem_sorted['desvio_p'].astype(int)
pontos_medios_fem_sorted
```

```python
#Calcular as diferenças ao quadrado
soma_diferencas_quad = pontos_medios_fem_sorted['desvio_p'].sum()
soma_diferencas_quad
```

-> A variânca é calculada com a soma das diferenças ao quadrado e a soma total de indivíduos da coluna 'counts'.

```python

#Dividir a soma pelo número total de idades

variancia_fem = soma_diferencas_quad / pontos_medios_fem_sorted['counts'].sum()
variancia_fem
```

```python
#Calcular o desvio padrão
desvio_padrao_fem = variancia_fem**0.5
desvio_padrao_fem
```

#####Describe

-> No métodos describe a variância é calculada também com as soma da diferença ao quadrado, porém o número divisor é 10 que reflete o número de faziaxa de idade

```python
#resumo estatístico da coluna faixa_etaria que está representada pelo ponto médio
resumo_fem = pontos_medios_fem_sorted['faixa_etaria'].describe()
resumo_fem
```

Obs: Neste caso o desvio padrão fica maior pois a divisão da soma da diferença ao quadrado não foi realizada com base no numero total de indivíduos.
